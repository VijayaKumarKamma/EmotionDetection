```python
count = 0
cap = cv2.VideoCapture(0)
frameRate = cap.get(5)
face_cascade = cv2.CascadeClassifier("C:\\Users\\kamma\\Anaconda3\\envs\\vdpnLeP35\\Library\\etc\\haarcascades\\haarcascade_frontalface_default.xml")
while True:
    frameId = cap.get(1) #current frame number--this is for frame rate
    ret, img = cap.read()
    if (ret != True):
        break
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,0),0)
        crop_img = img[y:y+h, x:x+w]
        if (frameId % (math.floor(frameRate/5)) == 0):
            filename ="Emotional_Imeges2/Training/Smile/frame%d.jpg" % count;count+=1
            #cv2.imwrite(filename, frame)
            print("saving in")
            cv2.imwrite(filename, crop_img)
            print("saving out")
            cv2.imshow('img',img)
            cv2.imshow('crop_img',crop_img)
    k= cv2.waitKey(1) & 0xff
    if k==27:
        break
cap.release()
print ("Done!")
```

```python
# Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Convolution2D(32, 3, 3, input_shape = (64, 64, 3), activation = 'relu'))
```

```python
classifier.add(MaxPooling2D(pool_size = (2, 2)))
```

```python
classifier.add(Convolution2D(32, 3, 3, activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))
```

```python
classifier.add(Flatten())
```

```python
classifier.add(Dense(output_dim = 128, activation = 'relu'))
classifier.add(Dense(output_dim = 1, activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory('Emotional_Imeges2/Training',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')

test_set = test_datagen.flow_from_directory('Emotional_Imeges2/Testing',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

```python
import PIL
```

```python
classifier.fit_generator(training_set,
                         samples_per_epoch = 5060,
                         epochs = 15,
                         validation_data = test_set,
                         nb_val_samples = 2292)
```

```python
# Part 3 - Making new predictions
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('photo.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = classifier.predict(test_image)
training_set.class_indices
if result[0][0] == 1:
    prediction = 'Smile'
else:
    prediction = 'Anger'
print(prediction)
```

```python
#Saving traind model to json and h5files.
# serialize model to JSON
#Save and load the model 
from keras.models import model_from_json
model_json = classifier.to_json()
with open("EmotionBinary2.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
classifier.save_weights("EmotionBinary2.h5")
print("Saved model to disk")
```

```python
#Later use
# load json and create model
json_file = open('EmotionBinary.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("EmotionBinary.h5")
print("Loaded model from disk")
```

```python
# Compiling and Making New Predictions

# compile loaded model on test data
loaded_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```
